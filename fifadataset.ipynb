```python
fifa_datasets["2021_dataset"]["BMI"] = fifa_datasets["2021_dataset"]["weight_kg"] / (fifa_datasets["2021_dataset"]["height_cm"] / 100) ** 2

one_hot = pd.get_dummies(fifa_datasets["2021_dataset"]['player_positions'])

fifa_datasets["2021_dataset"] = fifa_datasets["2021_dataset"].drop('player_positions',axis=1)

fifa_datasets["2021_dataset"] = fifa_datasets["2021_dataset"].join(one_hot)

label_encoder = preprocessing.LabelEncoder()
fifa_datasets["2021_dataset"]["preferred_foot"] = label_encoder.fit_transform(fifa_datasets["2021_dataset"]["preferred_foot"])

label_encoder = preprocessing.LabelEncoder()
fifa_datasets["2021_dataset"]["nationality"] = label_encoder.fit_transform(fifa_datasets["2021_dataset"]["nationality"])

keys = label_encoder.classes_
values = label_encoder.transform(label_encoder.classes_)
preferred_foots_dict = dict(zip(keys, values))
print(preferred_foots_dict)
```

```python
dataframes_to_be_merged=[]
for i in range(15, 22):
   fifa_datasets["20"+str(i)+"_dataset"]["Year"]="20"+str(i)
   dataframes_to_be_merged.append(fifa_datasets["20"+str(i)+"_dataset"])

fifa_dataset=pd.concat(dataframes_to_be_merged)
fifa_dataset=fifa_dataset.groupby(["short_name"]).mean()
fifa_dataset
```

```python
import numpy as np
```

```python
labels=["pace","shooting","passing","dribbling","defending", "overall"]
fifa_datset=fifa_dataset[labels]
num_vars = len(labels)

name="Cristiano Ronaldo"
values= fifa_datset.loc[name][labels].tolist()

angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()

values += values[:1]
angles += angles[:1]

fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

ax.set_theta_offset(np.pi / 2)
ax.set_theta_direction(-1)

ax.set_thetagrids(np.degrees(angles), labels)

ax.plot(angles, values, color='red', linewidth=1)

ax.fill(angles, values, color='red', alpha=0.25)

ax.set_title("Cristiano Ronaldo's Attribute Radar Plot")
```
